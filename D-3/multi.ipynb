```python
from sklearn.linear_model import LinearRegression

X = [
    [60, 22],
    [62, 25],
    [67, 24],
    [70, 20],
    [71, 15],
    [72,14],
    [75,14],
    [78,11]
]

y = [140,155,159,179,192,200,212,215] 

model = LinearRegression()
model.fit(X, y)

print("Model trained successfully!")

newx = float(input("Enter newx value "))
newy = int(input("Enter newy value "))
newy = model.predict([[newx, newy]])

print(f"\nnewy value is: {newy[0]:.2f} ")
2
```
